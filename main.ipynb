```python
import pandas as pd
import os
pd.set_option("display.precision", 7)
```

```python
# Read in individual csvs as dataframes

l = pd.read_csv('L.csv')
r = pd.read_csv('R.csv')
w = pd.read_csv('W.csv')
x = pd.read_csv('X.csv')
y = pd.read_csv('Y.csv')
```

```python
#Merge dataframesa and correct column names

df = pd.concat([l, r, w, x, y], axis=1)
cols = ['L', 'R', 'W', 'X', 'Y']
df.columns = cols
```

```python
# Remove all columns where R is 0

df.drop(df.loc[df['R']==0.0].index, inplace=True)
```

```python
# View first five rows
df.head()
```

```python
# Starting corner 1 X and Y

df['starting corner 1 [X]'] = df['X'] - df['L'] / 2
df['starting corner 1 [Y]'] = df['Y'] - df['W'] / 2
```

```python
# View starting corner 1 output

print('Starting Corner 1 [X]')
df['starting corner 1 [X]'].head()
print('Starting corner 1 [Y]')
df['starting corner 1 [Y]'].head()
```

```python
#Ending Corner 1 X, Y

df['ending corner 1 [X]'] = df['X'] + df['L'] / 2
df['ending corner 1 [Y]'] = df['Y'] + df['W'] / 2
```

```python
# View output of Ending Corner 1
print('Ending Corner 1 [X]')
df['ending corner 1 [X]'].head()
print('Ending corner 1 [Y]')
df['ending corner 1 [Y]'].head()
```

```python
#Second starting corner X, Y

df['starting corner 2 [X]'] = df['X'] - df['W'] / 2
df['starting corner 2 [Y]'] = df['Y'] - df['L'] / 2
```

```python
# View starting corner 2 output

print('Starting Corner 2 [X]')
df['starting corner 2 [X]'].head()
print('Starting corner 2 [Y]')
df['starting corner 2 [Y]'].head()
```

```python
# Second ending corner X, Y
df['ending corner 2 [X]'] = df['X'] + df['W']/2
df['ending corner 2 [Y]'] = df['Y']+ df['L'] / 2
```

```python
# View output of ending corner 2

print('Starting Corner 2 [X]')
df['ending corner 2 [X]'].head()
print('Starting corner 2 [Y]')
df['ending corner 2 [Y]'].head()
```

```python
# View a sample of the final output

#df['AutoCAD command'][0]
```

```python
# Convert to str
df[[col for col in df.columns]] = df[[col for col in df.columns]].astype(str)
```

```python
# Concat starting and ending corners back into single columns for convenience

df['starting corner'] = df['starting corner 1 [X]'] + ',' + df['starting corner 1 [Y]']
df['ending corner'] = df['ending corner 1 [X]'] + ',' + df['ending corner 1 [Y]']

df['starting corner.1'] = df['starting corner 2 [X]'] + ',' + df['starting corner 2 [Y]']
df['ending corner.1'] = df['ending corner 2 [X]'] + ',' + df['ending corner 2 [Y]']
df.head()
```

```python
# Create the formula

df['formulas'] = 'rectang ' + df['starting corner'] + ' ' + df['ending corner'] +'' + ' rectang ' + df['starting corner.1'] + ' ' + df['ending corner.1'] + ' circle ' + df['X'] + ',' + df['Y'] + ' ' + df['R']
```
